```python
#Ambil value dari key data dalam JSON, dan transformasi menjadi sebuah DataFrame.
import pandas as pd
import requests

# URL API yang ingin diambil data
url_api = "https://berita-indo-api-next.vercel.app/api/cnn-news/teknologi"

# Melakukan request ke URL API
response = requests.get(url_api)

# Mengubah respons menjadi JSON
data = response.json()

# Mengambil value dari key data
data_json = data['data']

# Mengubah JSON menjadi DataFrame
df = pd.DataFrame(data_json)

# Menampilkan DataFrame
df
```

```python
#Transformasi data kolom isoDate menjadi tipe data datetime
import pandas as pd

df['isoDate'] = pd.to_datetime(df['isoDate'])

print(df.dtypes)
```

```python
# Melakukan Aggregasi data berdasarkan kolom 'title', dikarenakan pada dataframe yang diperoleh tidak terdapat kolom "creator"
agg_df = df.groupby('title').size().reset_index(name='total_news')

# Mengurutkan berdasarkan jumlah berita terbanyak
sort_agg_df = agg_df.sort_values(by='total_news', ascending=False)

# Menampilkan hasil transformasi
print(sort_agg_df)
```
